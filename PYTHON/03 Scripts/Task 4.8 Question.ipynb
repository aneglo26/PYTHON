## 4.8.4

```python
# We start by creating a new collumn that we will call 'max_order'
combined_df_47['max_order'] = combined_df_47.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Then we test it to confirm it.
combined_df_47.head(60)
```

```python
# Finnaly we use loc() to create the loyalty flags
combined_df_47.loc[combined_df_47['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
combined_df_47.loc[(combined_df_47['max_order'] <= 40) & (combined_df_47['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
combined_df_47.loc[combined_df_47['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
combined_df_47['loyalty_flag'].value_counts(dropna = False)
```

## 4.8.5

```python
combined_df_47.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

As we can see, it does make sense that new customers tend to have a higher mean, as customers that buy frenquently mostly likely are re-purchasing products that they already use, while new customers are trying different products, which may result in more spending. 

## 4.8.6

```python
# We start by creating a new column showing the mean of prices
```

```python
combined_df_47['spend_mean'] = combined_df_47.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
combined_df_47
```

```python
# Then we create the conditions for the groups of High and low spender clients
```

```python

combined_df_47.loc[combined_df_47['spend_mean'] > 10, 'spender_type'] = 'High spender'
```

```python
combined_df_47.loc[combined_df_47['spend_mean'] <= 10, 'spender_type'] = 'Low spender'
```

```python
# And finally we count them
```

```python
combined_df_47['spender_type'].value_counts(dropna = False)
```

## 4.8.7 

```python
#Same as before, we start by creating a new column for the median of days_since_prior_order
```

```python
combined_df_47['avg_days_last_order'] = combined_df_47.groupby(['user_id'])['days_without_orders'].transform(np.median)
```

```python
# Then we create the groups
```

```python
combined_df_47.loc[combined_df_47['avg_days_last_order'] > 20, 'customer_frequency'] = 'Non-frequent customer'
```

```python
combined_df_47.loc[(combined_df_47['avg_days_last_order'] <= 20) & (combined_df_47['avg_days_last_order'] > 10), 'customer_frequency'] = 'Regular customer'
```

```python
combined_df_47.loc[combined_df_47['avg_days_last_order'] <= 10, 'customer_frequency'] = 'Frequent customer'
```

```python
# And finally count them again
```

```python
combined_df_47['customer_frequency'].value_counts(dropna = False)
```
